Sakuma piemers funkcijam Sin(x)-sarkana linija un Cos(x)-zila linija.

```python
import numpy
from numpy import * # Importē skaitlisko metožu bibliotēkas funkcijas
```

```python
x = linspace(0, 7, 70)  # Trešais arguments ir ģenerējamo elementu skaits
y = cos(x)
y2 = sin(x)

from matplotlib import pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funkcija $cos(x)$ un $sin(x)$')
plt.plot(x, y, color = "#112299") #blue for cos
plt.plot(x, y2, color = "#992211") #red for sin
plt.show()
```

```python
x = [0, 1, 2, 3]
y = [0, 2, 1, 2]
N = len(x)
legend = []
from matplotlib import pyplot as plt
#print(plt.__doc__)
# grafika ass iestatījumi
plt.axis([-1, 4, -1, 3])
# režģa ieslēgšana
plt.grid()
# x ass apzīmējums
plt.xlabel("$x$")
# y ass apzīmējums
plt.ylabel("$y$")
# grafika nosaukums
plt.title("$f(x)$ funkcijas grafiks")
# zīmēšana ar noklusētiem iestatījumiem
plt.plot(x,y)
legend.append("zīmēšana ar noklusētiem iestatījumiem")
# zīmēšana ar marķieriem
plt.plot(x[0],y[0],'gh',markersize = 10)
legend.append("zīmēšana ar marķieriem, 1. piemērs")
plt.plot(x[1:N],y[1:N],'ro',markersize = 5)
legend.append("zīmēšana ar marķieriem, 2. piemērs")
# leģendas (raksturlīķņu nosaukumu) pievienošana
#print(plt.legend.__doc__)
plt.legend(legend, loc = 3, fancybox=True, framealpha=0.5, facecolor="green")
# grafika publicēšana
plt.show()
```

Kods, kurs realize grafisko interpritaciju Sin(x) funkcijas izvirzijumiem rinda.

```python
x = linspace(0, 4, 70)  # Trešais arguments ir ģenerējamo elementu skaits
y = sin(x)



from matplotlib import pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funkcija $sin(x)$')
plt.plot(x, y, color = "#FF0000")   #red
y1 = x
plt.plot(x, y1, color = "#00FF00")  #green
y2 = y1 - x*x*x/(1*2*3)
plt.plot(x, y2, color = "#0000FF")  #blue
y3 = y2 + x*x*x*x*x/(1*2*3*4*5)
plt.plot(x, y3, color = "#8000FF")  #purple
y4 = y3 + x*x*x*x*x*x*x/(1*2*3*4*5*6*7)
plt.plot(x, y4, color = "#FFFF00")  #yellow
plt.show()
```

Kods, kurs realize grafisko interpritaciju Sin(sqrt(x)) * Sin(sqrt(x)) funkcijas izvirzijumiem rinda.

```python
x = linspace(0, 4, 70)  # Trešais arguments ir ģenerējamo elementu skaits
y = sin(sqrt(x)) * sin(sqrt(x))



from matplotlib import pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funkcija $sin(sqrt(x)) * sin(sqrt(x)$')
plt.plot(x, y, color = "#FF0000")   #red
y1 = x
plt.plot(x, y1, color = "#00FF00")  #green
y2 = y1 - x*x*x/(1*2*3)
plt.plot(x, y2, color = "#0000FF")  #blue
y3 = y2 + x*x*x*x*x/(1*2*3*4*5)
plt.plot(x, y3, color = "#8000FF")  #purple
y4 = y3 + x*x*x*x*x*x*x/(1*2*3*4*5*6*7)
plt.plot(x, y4, color = "#FFFF00")  #yellow
plt.show()
```
