### 3. Laboratorijas darbs (Class - 16 - Saknes meklesana)

```python
from math import sin, fabs, sqrt
from time import sleep
from numpy import *
from matplotlib import pyplot as plt
```

```python
def f(x):
    return sin(sqrt(x))*sin(sqrt(x))

k = 0
a = 1
b = 5

x = linspace(0 , 4 , 70)
y = sin(x*x)

plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funkcija $sin(sqrt(x))*sin(sqrt(x))$')
plt.plot(x, y, color = '#FF0000',linewidth = 1.0)
plt.show()

funa = f(a)
funb = f(b)

if (funa * funb > 0.0):
    print("Dotajā intervālā [%s, %s] sakņu nav"%(a,b))  
    sleep(1); exit()
else:
    print("Dotajā intervālā sakne(s) ir!")
deltax = 0.0001

while ( fabs(b-a) > deltax ):
    x = (a+b)/2; funx = f(x)
    if ( funa*funx < 0. ):
        b = x
    else:
        a = x

print("Sakne ir:", x)
print('Sakne ir: ',f(x))
print('Interāciju skaits k = ',k)
```
