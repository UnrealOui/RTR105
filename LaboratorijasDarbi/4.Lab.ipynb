### 4. Laboratorijas darbs (Class - 17 - Skaitliska diferencesana).

```python
from matplotlib import pyplot as plt
from numpy import sin, linspace, sqrt
```

```python
#diferencesana
def f(x):
    return sin(sqrt(x))*sin(sqrt(x))

x = linspace(0, 4, 10)
print(x)
y = f(x)
print(y)
#formulu izveide 1 atvasinajumam
firstder = [] 
deltax = x[1] - x[0]
N = len(x)

for i in range(N):
    temp = (f(x[i] + deltax) - f(x[i])) / deltax
    firstder.append(temp)
    
secondder = [] 

for i in range(N - 1):
    temp = (firstder[i + 1] - firstder[i]) / deltax
    secondder.append(temp)

legend = []


#atvasinajuma aprakstisana

legend = []
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")

plt.title("$sin(sqrt(x))*sin(sqrt(x))$ funkcija un tas atvasinājumi")

plt.plot(x, y, "m")
legend.append("$sin(sqrt(x))*sin(sqrt(x)) funkcija$")
plt.plot(x, y, "go")
legend.append("$sin(sqrt(x))*sin(sqrt(x)) funkcija(punkti)$")

plt.plot(x, firstder, "k")
legend.append("$sin(sqrt(x))*sin(sqrt(x))$ 1.k. atvasinājums")
plt.plot(x, firstder, "ro")
legend.append("$sin(sqrt(x))*sin(sqrt(x))$ 1.k. atvas.(punkti)")

plt.plot(x[0:len(x) - 1], secondder, "turquoise")
legend.append("$sin(sqrt(x))*sin(sqrt(x))$ 2.k. atvasinājums")
plt.plot(x[0:len(x) - 1], secondder, "ko")
legend.append("$sin(sqrt(x))*sin(sqrt(x))$ 2.k. atvas.(punkti)")

plt.legend(legend, loc=3) #Legendas lokacija attela

plt.show()
#Rezultats
```
